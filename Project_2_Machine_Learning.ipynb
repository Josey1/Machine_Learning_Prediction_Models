```python
# Looking at object columns and establishing their relevance
object_columns = Adult_df.select_dtypes(include=['object']).columns

# Loop through each object column and print value counts
for column in object_columns:
    print(f"Column: {column}")
    print(Adult_df[column].value_counts())
    print("\n")
```

 columns: native-country, occupation,and workclass have a number of missing values. I choose to drop these columns to reduce inconsistencies and inaccuracies in the analysis, as well as avoid complexities and bias in my data

```python
#Dropping the three columns from my dataset
columns_to_delete = ['native-country', 'occupation', 'workclass']

# Drop the specified columns from the DataFrame
Adult_df = Adult_df.drop(columns=columns_to_delete, axis=1)
```

```python
Adult_df.head()
```

```python
# Save a filter called "duplicated_rows" that identifies duplicated rows
duplicated_rows = Adult_df.duplicated()
duplicated_rows

# Get the sum of duplicated_rows filter to see the number of duplicate rows
Adult_df.duplicated().sum()
```

```python
# Drop duplicate rows from the DataFrame
Adult_df = Adult_df.drop_duplicates()
```

```python
# Plotting univariate visuals for all features except the target
for column in Adult_df.columns:
    if column != 'income':  # Skip the target variable 'income'
        if Adult_df[column].dtype == 'object':
            plt.figure(figsize=(8, 6))
            sns.countplot(x=column, data=Adult_df)
            plt.title(f'Univariate Visualization for {column}')
            plt.xticks(rotation=45)
            plt.show()
        else:
            plt.figure(figsize=(8, 6))
            sns.histplot(x=column, data=Adult_df, kde=True)
            plt.title(f'Univariate Visualization for {column}')
            plt.show()

# Plotting univariate visual for the target variable 'income'
plt.figure(figsize=(8, 6))
sns.countplot(x='income', data=Adult_df)
plt.title('Univariate Visualization for Target Variable (Income)')
plt.show()
```

```python
# Drop duplicate rows from the DataFrame
Wines_df = Wines_df.drop_duplicates()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the threshold for the number of unique categories to plot
top_n_categories = 10

# Plotting univariate visuals for all features except the target
for column in Wines_df.columns:
    if column != 'rating':  # Skip the target variable 'rating'
        if Wines_df[column].dtype == 'object':  # For categorical columns
            # Count the occurrences of each category
            category_counts = Wines_df[column].value_counts()
            # Select the top N categories
            top_categories = category_counts.head(top_n_categories)

            plt.figure(figsize=(8, 6))
            sns.barplot(x=top_categories.index, y=top_categories.values)
            plt.title(f'Top {top_n_categories} Categories for {column}')
            plt.xticks(rotation=45)
            plt.show()
        else:  # For numerical columns
            plt.figure(figsize=(8, 6))
            sns.histplot(x=column, data=Wines_df, kde=True)
            plt.title(f'Univariate Visualization for {column}')
            plt.show()

# Plotting univariate visual for the target variable 'rating'
plt.figure(figsize=(8, 6))
sns.countplot(x='rating', data=Wines_df)
plt.title('Univariate Visualization for Target Variable (rating)')
plt.show()
```
